```python
# Import the libraries as needed

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import thinkstats2
import thinkplot
import statsmodels.api as sm
```

```python
mj_trans = pd.read_csv("mj-clean.csv", parse_dates=[5])
mj_trans.head()
```

```python
# Function takes mj_trans DF and compute daily averages.
def GroupByDay(mj_trans, func=np.mean):
    grouped = mj_trans[["date", "ppg"]].groupby("date")
    daily = grouped.aggregate(func)

    daily["date"] = daily.index
    start = daily.date[0]
    one_year = np.timedelta64(1, "Y")
    daily["years"] = (daily.date - start) / one_year

    return daily
```

```python
# Calculate daily averages using the GroupByDay function
daily_averages = GroupByDay(mj_trans)
daily_averages.head()
```

```python
# Definiton for Quadratic Model by Daily
def RunQuadraticModelByDaily(daily_averages):
    # Extract years and prices from the daily data
    years = daily_averages['years']
    prices = daily_averages['ppg']

    # Fit a quadratic model
    quad_fit = np.polyfit(years, prices, deg=2)

    # Generate predictions using the quadratic model
    predictions = np.polyval(quad_fit, years)
    
    # Calculate residuals
    residuals = prices - predictions
    
    # Perform OLS regression
    X = sm.add_constant(years)
    model = sm.OLS(prices, X).fit()
    
    # Print regression summary
    print(model.summary())
    
    # Scatterplot of actual prices
    plt.figure(figsize=(18, 10))
    plt.scatter(years, prices, label='Actual Prices')
    plt.plot(years, np.polyval(quad_fit, years), label='Fitted Values', linestyle='dotted')
    plt.xlabel('Years')
    plt.ylabel('Price per gram ($)')
    plt.legend()
    plt.title('Actual Prices and Fitted Values')

    # Coefficient plot
    plt.figure(figsize=(18, 10))
    coef_labels = ['Constant', 'Years']
    coef_values = [model.params[0], model.params[1]]
    plt.bar(coef_labels, coef_values)
    plt.ylabel('Coefficient Value')
    plt.title('Regression Coefficients')
    plt.tight_layout()
    plt.show()

    # Scatterplot for residuals and predictions
    plt.figure(figsize=(18, 10))
    plt.scatter(years, residuals, label='Residuals', color='purple')
    plt.scatter(years, predictions, label='Quadratic Model Predictions', color='red')
    plt.xlabel('Years')
    plt.ylabel('Price per gram ($)')
    plt.legend()
    plt.title('Scatterplot of Residuals and Quadratic Model Predictions')
    plt.show()

    # Scatterplot of actual prices vs fitted vlues
    plt.figure(figsize=(18, 10))
    plt.scatter(years, prices, label='Actual Prices')
    plt.plot(years, np.polyval(quad_fit, years), label='Fitted Values', linestyle='dotted', linewidth=6, color='orange')
    plt.xlabel('Years')
    plt.ylabel('Price per gram ($)')
    plt.legend()
    plt.title('Actual Prices and Fitted Values')
    plt.show()

    # Plot the original prices, fitted values, and the quadratic model predictions
    plt.figure(figsize=(18, 10))
    plt.plot(daily_averages['years'], prices, label='Original Daily Averages')
    plt.plot(daily_averages['years'], np.polyval(quad_fit, years), label='Fitted Values', linestyle='dotted')
    plt.plot(daily_averages['years'], predictions, label='Quadratic Model Predictions', linestyle='dashed')
    plt.xlabel('years')
    plt.ylabel('Price per gram ($)')
    plt.legend()
    plt.title('Quadratic Model for Daily Average Prices')
    plt.show()
    
    # Plot the scatterplot of actual prices and predictions
    plt.figure(figsize=(18, 10))
    plt.scatter(years, prices, label='Actual Prices', color='blue')
    plt.scatter(years, predictions, label='Quadratic Model Predictions', color='red')
    plt.xlabel('Years')
    plt.ylabel('Price per gram ($)')
    plt.legend()
    plt.title('Scatterplot of Actual Prices and Quadratic Model Predictions')
    plt.show()

    # Calculate RMSE
    mean_error = np.mean(prices - predictions)
    rmse = np.sqrt(np.mean((prices - predictions)**2))
    print(f"Mean Error: {mean_error}")
    print(f"Root Mean Squared Error: {rmse}")
    
    return quad_fit, residuals

# 
RunQuadraticModelByDaily(daily_averages)
```

- The above implementation employs a quadratic regression model to predict daily average prices over time. Utilizing the statsmodels library, the code performs Ordinary Least Squares (OLS) regression, offering insights into the model's performance through coefficient analysis and relevant statistics. The visualizations, including scatterplots of actual prices against predictions and residual analysis, help assess the accuracy of the model and its ability to capture trends and patterns in the dataset.

```python
# Permutation-based test for serial correlation
class SerialCorrelationTest(thinkstats2.HypothesisTest):

    # Test statistic
    def TestStatistic(self, data):
        series, lag = data
        test_statistic = abs(thinkstats2.SerialCorr(series, lag))
        return test_statistic
    
    # Run the model of the null hypothesis
    def RunModel(self):
        series, lag = self.data
        permutation = series.reindex(np.random.permutation(series.index))
        return permutation, lag
```

```python
# Test the serial correlation for consecutive prices
test = SerialCorrelationTest((price, 1))
p_value = test.PValue()
print("Correlation between consecutive prices")
print("Serial Correlation:", test.actual)
print("P-value:", p_value)
print()
```
