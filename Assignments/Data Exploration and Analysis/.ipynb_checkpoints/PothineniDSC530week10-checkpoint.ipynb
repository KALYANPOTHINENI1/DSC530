```python
# Define a function to create a survival function from a cumulative distribution function (CDF)
def MakeSurvivalFromCdf(cdf, label=""):
    ts = cdf.xs
    ss = 1 - cdf.ps
    return survival.SurvivalFunction(ts, ss, label)
```

```python
# Read respondent data from different years
resp5 = survival.ReadFemResp1995()
resp6 = survival.ReadFemResp2002()
resp7 = survival.ReadFemResp2010()
```

```python
# Define a function to clean respondent data
def CleanData(resp):
    resp.cmdivorcx.replace([9998, 9999], np.nan, inplace=True)
    resp["notdivorced"] = resp.cmdivorcx.isnull().astype(int)
    resp["duration"] = (resp.cmdivorcx - resp.cmmarrhx) / 12.0
    resp["durationsofar"] = (resp.cmintvw - resp.cmmarrhx) / 12.0

    month0 = pd.to_datetime("1899-12-15")
    dates = [month0 + pd.DateOffset(months=cm) for cm in resp.cmbirth]
    resp["decade"] = (pd.DatetimeIndex(dates).year - 1900) // 10
```

```python
# Define a function to clean respondent data
def clean_data(resp):
    # Replace specific values with NaN in cmdivorcx column
    resp['cmdivorcx'] = resp['cmdivorcx'].replace([9998, 9999], np.nan)
    
    # Create a binary column 'notdivorced' based on whether cmdivorcx is null
    resp['notdivorced'] = resp['cmdivorcx'].isnull().astype(int)
    
    # Calculate duration of marriage and duration so far columns
    resp['duration'] = (resp['cmdivorcx'] - resp['cmmarrhx']) / 12.0
    resp['durationsofar'] = (resp['cmintvw'] - resp['cmmarrhx']) / 12.0

    # Convert century-months of birth to decades
    birth_date_reference = pd.to_datetime("1899-12-15")
    birth_dates = [birth_date_reference + pd.DateOffset(months=cm) for cm in resp['cmbirth']]
    resp['decade'] = (pd.DatetimeIndex(birth_dates).year - 1900) // 10
```

```python
# Clean and process respondent data for different years
CleanData(resp6)
#resp6.head()
married6 = resp6[resp6.evrmarry == 1]
CleanData(resp7)
married7 = resp7[resp7.evrmarry == 1]
```

```python
# Define a function to estimate survival by decade
def estimate_survival_by_decade(groups, **options):
    num_groups = len(groups)
    thinkplot.PrePlot(num_groups)

    for name, group in groups:
        _, survival_function = estimate_survival(group)
        thinkplot.Plot(survival_function, **options)

    thinkplot.Config(xlabel="Years", ylabel="Fraction undivorced", axis=[0, 28, 0, 1])
```

```python
# Define a function to estimate survival curves
def estimate_survival(resp):
    # Separate rows where respondents are not divorced (still married)
    not_divorced = resp[resp.notdivorced == 1]
    # Extract durations for ongoing marriages
    complete_duration = not_divorced.durationsofar.dropna()

    # Separate rows where respondents are divorced
    divorced = resp[resp.notdivorced == 0]
    # Extract durations for completed marriages
    ongoing_duration = divorced.duration.dropna()

    # Estimate hazard function using durations of completed and ongoing marriages
    hazard_func = survival.EstimateHazardFunction(complete_duration, ongoing_duration)
    # Convert hazard function to survival function
    survival_func = hazard_func.MakeSurvival()

    # Return both the hazard function and survival function
    return hazard_func, survival_func

```

```python
# Define a function to resample and plot divorce curves by decade
def ResampleDivorceCurveByDecade(resps):
    for i in range(41):
        # Resample rows from each respondent and collect into samples
        samples = [thinkstats2.ResampleRowsWeighted(resp) for resp in resps]
        
        # Combine all samples into a single DataFrame
        sample = pd.concat(samples, ignore_index=True)
        
        # Group the samples by decade of birth
        groups = sample.groupby("decade")
        
        # Add labels to the first group for better visualization
        if i == 0:
            survival.AddLabelsByDecade(groups, alpha=0.7)

        # Estimate and plot survival curves for each group (decade)
        estimate_survival_by_decade(groups, alpha=0.1)
        
    thinkplot.Config(xlabel="Years", ylabel="Fraction undivorced", axis=[0, 28, 0, 1])
```

```python
# Resample and plot divorce curves by decade for multiple years
ResampleDivorceCurveByDecade([married6, married7])
```
